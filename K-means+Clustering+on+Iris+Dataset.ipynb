```python
from sklearn import datasets
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import numpy as np

%matplotlib inline
```

```python
iris = datasets.load_iris()
X = iris.data
df = pd.DataFrame(X)
df.columns = iris.feature_names
```

The dataset we're using today contains information about flowers called an iris.  We're going to use K-Means clustering to atttempt to identify unique types of iris.

Step 1:  Use my method below to identify how many groups are in the dataset.

Step 2:  add the kmeans cluster labels into the dataframe df.


```python
df.head()
```

```python
def find_k (X, k_range, sample_percent=1):
    """
    k_range: a list of possible k values
    X, the data we're clustering on
    """
    from sklearn.cluster import KMeans
    import matplotlib.pyplot as plt
    import numpy as np
    from scipy.spatial.distance import cdist
    from sklearn.metrics import pairwise_distances

    N = X.shape[0]
    sampleSize = X.shape[0] * sample_percent

    if sampleSize > 0:
        index = np.arange(np.shape(X)[0])
        np.random.shuffle(index)
        X =  X[index, :]


    mean_distortions=[]
    for k in k_range:
        #cluster using k, then calculate the mean distortion (average distance to closest centroid)
        kmeans_model = KMeans(n_clusters=k, init='k-means++', n_jobs=-1).fit(X)
        mean_distortions.append(sum(np.min(pairwise_distances(X, kmeans_model.cluster_centers_,
                                                              metric='euclidean'),axis=1)) / X.shape[0])


    #visualize results
    plt.plot(k_range, mean_distortions)
    plt.xlabel("K Value")
    plt.ylabel("Mean Distortion")
    plt.title("Elbow Graph for Mean Distortion per K")
    plt.show()
```

```python
# find mean distortion for K=1 to 9
find_k(np.matrix(X), range(1,10), 1)
```

### Finding the Elbow
Why?: Using K Values between 1 and 9, we graphically see pronounced elbows in our results at 2 and 3.  Although K=2 represents a significant delta in mean distortion from K=1 in relative terms, I don't think we can ignor the 25% of mean distortion from K=2 to K=3, thus I've deduced that 3 groups is more appropriate

However, regardless of if you go with 2 or 3, this seems like a too small for different types of irises.  Let's try scaling the values and see if normalization helps.

```python
# feature scaling and re-run function find_k
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
find_k(np.matrix(X_scaled), range(1,10), 1)
```

Not much difference graphically so we stick with K=3

```python
kmeans_model = KMeans(n_clusters=3, random_state=30)
kmeans_model.fit(X_scaled)
```

```python
df['labels'] = kmeans_model.labels_
df.head()
```

```python
print('Label 0:', df[df['labels'] == 0].shape[0])
print('Label 1:', df[df['labels'] == 1].shape[0])
print('Label 2:', df[df['labels'] == 2].shape[0])
```

```python
from sklearn.manifold import TSNE
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

model = TSNE(n_components=3, learning_rate=12, random_state=30)
X_tsne = model.fit_transform(X)
plt.figure(figsize=(10, 10))
ax = plt.subplot(111, projection='3d')
ax.scatter(X_tsne[:,0], X_tsne[:, 1], X_tsne[:, 2], c=kmeans_model.labels_)
```

### Final Analysis
Based on our K-means model with K=3, we've identified all 150 observations in to one of three groups: 0, 1 or 3.  The distribution across each of the groups was fairly even with about one-third of the observations going to each cluster.  The plot above also shows a nice grouping with logical clusters.
